## Pandas (Fast)

```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_csv('users.csv', index_col='id')
data.head()
```

```python
data.iloc[[4, 5, 6]]
```

### Work

```python
# Usuarios que tienen una edad mayor a 40
data[data.edad > 40]
```

```python
data[(data.edad > 40) & (data.genero == 'female')]
```

### Excluir usuarios que tengan correo example.com

```python
data[data.email.str.endswith('@example.com')]
```

```python
data[~data.email.str.endswith('@example.com')]
```

### Filtrar a los hombres que viven en canada

```python
data[(data.genero == 'male') & (data.pais == 'Canada')]
```

## Agrupar por sexo

```python
data.groupby('genero')
```

```python
data.groupby('genero').count()
```

### Mostrar el pais con mas mujeres

```python
data[data['genero']=='female'].groupby('pais')['pais'].count().nlargest(1)
```
